# ej 1


Construir arrays a partir de una lista simple
y una lista anidada usando tipos de datos
numéricos distintos para cada una.


```python
import numpy as np
```

```python
# listas con distintos tipos numéricos
lista_simple = [1, 2, 3]                    # enteros
lista_anidada = [[1.0, 2.0], [3.0, 4.0]]    # flotantes

array = np.array(lista_simple)
print(array)
print("-------------")
array = np.array(lista_anidada)
print(array)
```

# Ejercicio 2
Dado un vector de pesos(kg) y otro vector
alturas(metros), calcular el índice de masa
corporal general (un vector imc).


```python
pesos = np.array([50,65.5,89.2,100])
alturas = np.array([1.6,1.7,1.8,1.75])
alturas_imc = np.square(alturas)

print (pesos/alturas_imc)

```
