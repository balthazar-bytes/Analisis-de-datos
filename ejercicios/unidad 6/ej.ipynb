## Ejercicio 1
Construir 3 gráficos de línea a partir de un array
creado con la función linspace de numpy.
1. Uno que represente una curva lineal.
2. Otro que represente una curva cuadrática.
3. Un tercero que represente una curva
logarítmica.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(3, 1)
x = np.array([1,2,3,4,5])
axs[0].plot(x)
axs[1].plot(x**2)
axs[2].plot(np.log(x))
```
