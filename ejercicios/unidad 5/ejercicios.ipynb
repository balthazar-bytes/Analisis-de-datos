3. Edad promedio por clase de los que

```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
ruta = r"C:\Users\balti\OneDrive\Desktop\programacion\Analisis-de-datos\teoria\modulo 4\csv\titanic.csv"
df = pd.read_csv(ruta)
df
```

```python
df[['Survived','Age']].groupby('Survived').mean(numeric_only=True)
```

```python
df[["Survived",'Age']].groupby('Survived').agg(['min','max']).rename(index= {0: 'muerto',1:'superviviente'})
```

```python
df[['Survived','Pclass','Age']].groupby(['Survived','Pclass']).mean(numeric_only=True)
```

2. Hacer un gráfico del tipo histograma que represente la frecuencia de edades únicamente de las

```python
df[['Survived','Age']].groupby('Survived').mean().rename(index={0:'muerto',1:'superviviente'}).plot.bar(title="edad promedio de supervivientes")
```

```python
df.loc[(df['Sex'] == 'female') & (df['Survived'] == 1), 'Age'].plot.hist(bins=20, title='Frecuencia de edades de mujeres sobrevivientes')
```
