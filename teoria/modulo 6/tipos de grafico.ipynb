## graficos


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## grafico d puntos

```python
x,y = np.random.random((2,30))
c = np.random.randint(0,9,30)
s = np.random.randint(10,150,30)

plt.scatter(x,y,c=c,s=s) #permite graficr puntos en un plano
```

## historigrama


```python
x1 = np.random.normal(4,3,1000)
x2 = np.random.normal(1,1,1000)
plt.hist((x1,x2),label=["grupo 1","grupo 2"])
plt.title("histograma")
plt.legend()
```
