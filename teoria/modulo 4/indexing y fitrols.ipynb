## label indexing
pandas provee un metodo para seleccionar elementos en base a las etiquetas atraves de __.loc__ , cuando se usa el .loc los numero representaran las etiquetas y no las posiciones, tambien se puede hacer slice con etiquetas

```python
data.loc[90,'Name'] # esto muesta el nombre numero 90
```

```python
data.loc[25:30,'PassengerId':'Pclass'] # esto es un slice, muestra desde el 25 al 30 
```

```python
data.loc[[25,30,200],['Survived','Age']] # aca me muiesta solo el 25 y el 30, los del medio no importan
```

## integer indexing

no solo existe el indexing .loc sino que tambien el __.iloc__ para acceder a elementos segun la posiciones, en este caso los numeros representan las posiciones y no las etiquetas

```python
data
```

```python
data.iloc[125,4] # el primer numero representa que numero de posiciones quiero y el segundo que etiqueta quiero (columna), aca me muestra solo 1 dato
```

```python
data.iloc[10:18,-3:] # esto me muestra los ultimos 3 por el negativo, pero tambien podria hacer que me muestre los primero tres
```

```python
data.iloc[10:18,:3] # este me muestra los primeros 3
```

```python
data.iloc[[25,30],[2,5]] # selecciono cuales ilas quiero y que quiero de essa filas
```

## FILTROS

tambien es posible seleccionar elementos segun condiciones de sus valores, la expresion condicional produce un arreglo booleano y devuleve los valores de ese array que sean verdadesro

```python
data[data['Age'] >= 30] # este data es como que diga 'data["mayores de 30"]'
```

```python
data.loc[data['Age'] >= 30, ['Name',"Sex"]] # aca se combina loc con filtros pero lo que se hace es filtrar y despues los que cumplan eso solo se muestras lo que sea de interes, en este caso su nombre y su sexo
```
