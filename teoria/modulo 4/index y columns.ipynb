## indices 
las etiquetas son una parte fundamental de las estructuras de pandas, muchas funcionalidades dependen de ellas
Por lo tanto, es importante poder
manipular y modificarlas según sea necesario. Las
etiquetas están alojadas en una estructura llamada
__Index__

```python
import pandas as pd
import numpy as np
```

```python
array = np.array([
    [3,6,2],
    [6,8,5],
    [7,0,5],
    [4,3,8]
])
df = pd.DataFrame(array)
df
```

```python
df.index = ['fila 1','fila 2','fila 3','fila 4']
df.columns= ['col 1','col 2','col 3']
df
```

Los objetos de tipo Index no soportan la asignación de ítems, no
se puede cambiar uno de sus elementos por asignación directa.
Esta modificación la podemos realizar de dos maneras:
* Una es modificando directamente el array de numpy que aloja los datos del Index a través del atributo values.
* La otra forma es usando el método rename

```python
df.index.values[0] = 'FILA 1'
df
```

```python
df.rename(index= {'fila 2': "nueva fila 2"},inplace=True)
df
```

```python
df.rename(columns= lambda x: x.replace('col','columna'),inplace=True)
df
```

## reset_index
para convertir el index en una columna se usa el metodo reste_index, este transforma el index en una columna y resetea el indece a un rango de numeros

```python
df.reset_index()
```

## reindex
El método reindex permite cambiar el orden
en que se encuentran las etiquetas pero
manteniendo la alineación de los datos con la
etiqueta original. Sólo se conservan las filas
cuyas etiquetas se indiquen explícitamente.

```python
df.reindex(['marzo','enero','septiebre'])
```
