# pandas
pandas es un modulo que esta echo sobre numpy
pandas permite trabajar con distintos tipos de datos para cada columna, __es semejante a una base de datos__

hay dos estructuras de datos: __serie y dataframe__ 

## series

es una estructura de dats unidimencional donde cada dato tiene ademas una etiqueta. es una estructura __mutable__ a la que no solo se le puede cambiar sus elementos, sino que tmb quitar o agregar

a las etiquetas se las conoce como __index__ y las etiquetas no tiene pq ser unicas
estas etiquets agregan funcionalidad para manipular los datos (se usa __pd.series()__)

```python
import  pandas as pd
pd.Series([5,7,2])
```

```python
pd.Series([5,7,2],index=['a','b','c'], dtype= 'int8',name='numeros')
```

```python
pd.Series({'martin':8,'juan':9,'lucia':10},name='notas')
```

## dataframe

es una estructura bidimensionbal  de tipo tabla con etiquetas para las filas  columnas, __cada columna puede tener su propio tipo de dato__

es mutable tanto por sus datos como por su tamano, cada columna es una serie dond todas las series tiene el mismo index

se pueden construir dataframes apartir de muchas estructuras dist, como listas, array y diccionarios

```python
pd.DataFrame([
    [0,1,2],
    [3,4,5],
    [6,7,8]
])
```

```python
pd.DataFrame([
    [0,1,2],
    [3,4,5],
    [6,7,8]
],index=['a','b','c'],columns=['lunes','martes','miercoles'])
```

```python
import numpy as np
```

```python
array = np.random.randint(0,50,(3,5))
pd.DataFrame(array)
```

```python
pd.DataFrame({'Columna 1':[1,2,3],'columna 2':[4,5,6]},index=['A','B','C'])
```
