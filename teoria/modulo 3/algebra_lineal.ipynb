# algebra


## producto escalar
es multiplicar cada elemento o todos por un escalar


```python
import numpy as np
a = np.array([1,2,3])

b = np.array([4,5,6])

print ( a@b) # devuelve solo un numero
```

## producto X o producto matricial
es la multiplicacion de fila X columna, el numero de columnas de la primeraa matriz debe de coincidir con el numero de filas de la segunda
ejemplo: 2x3 3x2 ( si se puede hacer)

```python
a = np.array([
    [2,5,3],
    [7,1,4]
])

b = np.array([
    [3,2],
    [9,6],
    [5,8]
    ])

a.dot(b)
```

## determinante

tiene que ser una matriz cuadrada, 3x3,4x4,2x2,10x10,etc

```python
a = np.array([
    [1,2,3],
    [5,6,7],
    [3,4,2]
])
print(np.linalg.det(a))
```
