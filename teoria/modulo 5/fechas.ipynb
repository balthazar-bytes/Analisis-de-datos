Cuando tenemos datos relacionados a fechas, podemos
utilizar la serie de tiempo como índice. Esto le da mucha
funcionalidad ya que podemos seleccionar datos
refiriéndonos a las fechas.



Para utilizar la columna de fechas como índice podemos
usar set_index. En este caso pandas crea un tipo de
dato especial, el DatetimeIndex. Como sólo se trata de
fechas podemos acceder a los métodos directamente,
sin pasar por el atributo dt.



Ahora podemos seleccionar los datos según criterios
temporales, como todos los datos de un año o de todos
los lunes, usando indexing y filtros.

```python
data.set_index("Date",inplace=True)
data
```
