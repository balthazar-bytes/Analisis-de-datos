## matplotlib

es el principal modulo de graficos de python, podemos acceder a las funcionalidades para graficar atraves del metodo __plot__
ej:
* plt.bar: grafico de barras
* plt.hist: historigarma
* plt.pie: torta grafica


```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
ruta = r"C:\Users\balti\OneDrive\Desktop\programacion\Analisis-de-datos\teoria\modulo 4\csv\titanic.csv"
data = pd.read_csv(ruta)
data
```

## histograma
es una representacion de la distribuccion de un conjunto de numeros, el rango se divide en barras y la altira de la barra es proporcional a la cantidad de datos

```python
data['Age'].plot.hist()
# la altura muestra la frecuencia de la edad, aca vemos que hay mayormente entre 20 a 30 anios
```

```python
data['Age'].plot.hist(bins=20, title='Distribucon de edades',legend=True,color='red')
```

## grafico de barras
este grafico muestra altura variable. generalmente son utiles para mostrar estadistica tomadas __por categoria por medio de groupby__

```python
data.groupby('Sex').size().plot.bar()
```

```python
data.groupby('Sex')['Survived'].sum().plot.bar()
```

## grafico torta


```python
data.groupby('Sex').size().plot.pie(title='Proporcion de hombre y mujeres')
```

```python
data.groupby('Sex')['Survived'].sum().plot.pie(title='Proporcion de hombre y mujeres entre los sobevivientes')
```
