## strings 


Existen numerosos métodos para trabajar con strings en objetos tipo Series e Index (no aplica a Dataframes).


En general, tienen el mismo nombre que los métodos de
python para strings y están agrupados en el atributo
str. Así es posible hacer operaciones sobre todos los
elementos de la estructura con un único comando.



Por ejemplo, para cambiar entre mayúsculas y
minúsculas existen lower, upper, title y swapcase.


```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
data["Name"].str.lower()
```

```python
data["Name"].str.upper()
```

__split__ permite separar el strings por partes

por defecto usa los espacios en blanco pero podemos definirlo nosotros

```python
data["Name"].str.split()
```

__count()__ cuenta cuantaas veces aparece un substring en cada elemento

```python
data["Name"].str.count("Mr.")
```

__Contains__ devuelve True o False segun cada string qe contenga o no cierto substring

```python
data["Name"].str.contains("Mr.")
```

__startswith y endswitch__ permite saber si cada string empieza o termina con cierto sufijo o prefijo


```python
data["Name"].str.startswith("Andersson").sum()
```
