## groupby

es facil hace restadisticas pero a veces estamos interesados en realizar estadisticas en subconjuntos segun la categoria a la que pertenezaca, por ejemplo queremos saber __la edad promedio para hombre y mujeres__ 

```python
import pandas as pd
import numpy as np
```

```python
ruta = r"C:\Users\balti\OneDrive\Desktop\programacion\Analisis-de-datos\teoria\modulo 4\csv\titanic.csv"
df = pd.read_csv(ruta)
df
```

usamos grupoby para agrupar los datos segun valores de una o mas clumnas y los vamos a combinar con algun metodo

```python
df.groupby('Sex').count()
```

```python
df.groupby('Sex').mean(numeric_only=True)
```

si queremos restringir el resultado de ciertas columnas podemos hacerlo de varias formas: 
* un es seleccionando la columna que nos interesa operacionar
* otra es seleccionar las columans nesesarias antes del groupby
* y la ultima es seleccionar la columna de interes dsp de hacer el group

```python
# ej 1
df.groupby("Sex").mean(numeric_only=True)['Age']
```

```python
#ej 2
df[['Survived','Age','Sex']].groupby('Survived').mean(numeric_only=True)
```

```python
# ej 3
df.groupby('Sex')['Age'].mean(numeric_only=True)
```

```python
#promedio de edad segun el sexo
df.groupby('Sex')[['Sex','Age']].mean(numeric_only=True)
```

```python
#maximo de edad en cada clase
df.groupby('Pclass')[['Age','Pclass']].max()
df[['Age','Pclass',]].groupby('Pclass').max()
```

En caso de querer realizar varias operaciones a los mismos
grupos se puede utilizar aggregate (o su alias, agg).

```python
df[['Age','Pclass']].groupby('Pclass').agg(['min','max','mean','std'])
```
