# funciones anonimas


tambien llamadas lambdas, nos permite hacer peque;as funciones sin nombres en el momento, sin necesidad de definirlas

estan limitadas a solo una exprecios

```python
lambda x,y: x+y #esto equivalo a una funcion suma 

lambda x: x%2==0 # funcion par

lambda x: x**2+25
```

estas funciones se usan mayormente para una craecion rapida y corta , y no tener que estar definiendo cada 2x3 una nueva funcion

### Ejemplo
Supongamos que tenemos una lista de strings que
contienen nombres pero están todos en formato distinto
e incluso algunos tienen espacios en blanco que sobran.
Queremos que queden todos igual y para eso tenemos
que aplicar primero el método strip y luego capitalize
a cada elemento. Podemos hacer uso de map y lambda
para realizar esto de forma sencilla.

```python
nombres = ['Juan',"cAroLina","RaUL", "    FlaViA"]

list(map(lambda x: x.strip().capitalize(),nombres))
```
