# FUNCIONES DE ORDEN SUPERIOR

son aquellas que toman otras funciones como parametro o devuelven una funcion como return


## MAP()

la funcion de orden superior map() sirve para poder pasar como parametro una funcion 

```python
def cuenta(x):
    return x**2 + 25
numeros = [8,2,3,5,1,3]
map(cuenta,numeros) # <map at 0x1bd048f97b0>, esto pasa pq esta funcion retorna un generado, esto significa que sus  elementos no son calculados hasta que se requiera

list(map(cuenta,numeros))
```

## FILTER( )

esta funcion permite seleccionar elementos de un iterable usando otra funcion para decidir que elementos son seleccionados

al igua lque map devuelve un generrador

```python
numeros = [1,2,3,5,1,0,5,4,3,6]

def par(x):
    return x%2 == 0

list(filter(par,numeros)) # filter lo que hace es seleccionar a los resultados true y usa los numero que hicieron que den true
```
